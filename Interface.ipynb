```python
def log_file_write(text):
    log_string = text + str(datetime.datetime.now()) + "\n"
    f = open("log/log_file.txt", "r+")
    f.seek(0, 2)
    f.write(log_string)
    f.close()
    #print(f.tell())
```

```python
def flashing(label, im1, im2):
    while True:
        label.image = im1
        time.sleep(0.2)
        label.image = im2
        time.sleep(0.2)
```
