```python
import openpyxl
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import random
import time
import warnings

wb = openpyxl.load_workbook(r"data_sheets\EK3.xlsx")
wb1 = openpyxl.load_workbook(r"data_sheets\Книга1.xlsx")
sheet = wb['Отчет']
sheet1 = wb1['Лист4']
warnings.filterwarnings('ignore')
```

```python
def read_value(temp, i, n, col):
    temp.append(sheet.cell(row = i, column = col).value)
    if (temp[i-n] == ''):
        temp[i-n] = temp[i-n-1]
    #print(temp[i-n])
    try:
        temp[i-n] = float(temp[i-n])
    except:
        return 0
```

```python
def averager(array):
    for i in range (1, len(array)-1):
        array[i] = ((array[i-1] + array[i+1])/ 2) 
```

```python
temp1 = []
temp2 = []
flow1 = []
temp3 = []
temp4 = []
temp5 = []
flow2 = []
pres1 = []
n = 18
num = 30549
for i in range (n, num):
    read_value(temp1, i, n, 2)
    read_value(temp2, i, n, 3)
    read_value(flow1, i, n, 4)
    read_value(temp3, i, n, 5)
    read_value(temp4, i, n, 6)
    read_value(temp5, i, n, 7)
    read_value(flow2, i, n, 8)
    read_value(pres1, i, n, 9)
    
plt.plot(pres1)
plt.show()

averager(temp1)
averager(temp2)
averager(flow1)
averager(temp3)
averager(temp4)
averager(temp5)
averager(flow2)
#averager(pres1)

plt.plot(temp1)
plt.show()
i = 1
for i in range (1, num-n):
    sheet1.cell(row = i, column = 1).value = temp1[i]
    sheet1.cell(row = i, column = 2).value = temp2[i]
    sheet1.cell(row = i, column = 3).value = flow1[i]
    sheet1.cell(row = i, column = 4).value = temp3[i]
    sheet1.cell(row = i, column = 5).value = temp4[i]
    sheet1.cell(row = i, column = 6).value = temp5[i]
    sheet1.cell(row = i, column = 7).value = flow2[i]
    sheet1.cell(row = i, column = 9).value = pres1[i]
wb1.save(r"data_sheets\Книга1.xlsx")
```
